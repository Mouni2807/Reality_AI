```python
df = df.dropna(thresh=df.shape[0]*0.5, axis=1)

# Fill numeric NaN with median
num_cols = df.select_dtypes(include=['int64','float64']).columns
for col in num_cols:
    df[col] = df[col].fillna(df[col].median())

# Fill categorical NaN with mode
cat_cols = df.select_dtypes(include=['object']).columns
for col in cat_cols:
    df[col] = df[col].fillna(df[col].mode()[0])

print("Missing values after filling:", df.isnull().sum().sum())
```

```python
# ===================================
# Prophet Forecasting - Zillow Trends
# ===================================

from prophet import Prophet

# The 'cleaned_df' is used here, ensuring 'Date' and 'ZHVI_AllHomes' are available
# and not scaled/dropped.
# Make sure to run the previous cells, including the updated In[13] logic
# and In[12] which defines cleaned_df.

# Pick a specific region for forecasting
region = "park_citysedgwickks "  # This region was confirmed to have 261 non-NaN values
# Ensure 'Date' is a datetime object in cleaned_df before filtering/dropna
if cleaned_df['Date'].dtype != '<M8[ns]':
    cleaned_df['Date'] = pd.to_datetime(cleaned_df['Date'])

# Filter data for the specific region
# Use .dropna() to remove NaNs before fitting Prophet
# List all columns you need: date ('Date'), target ('ZHVI_AllHomes'), and all regressors
required_columns = ["Date", "ZHVI_AllHomes", 
                    "ZHVI_SingleFamilyResidence", 
                    "ZHVIPerSqft_AllHomes", 
                    "ZHVI_3bedroom", 
                    "ZHVI_4bedroom",
                    'ZHVI_MiddleTier',"ZHVI_TopTier",'ZHVI_BottomTier']

# Filter the data, including ALL necessary columns
region_df = cleaned_df[cleaned_df["RegionName"].str.strip() == region.strip()][required_columns].dropna()

# Check if there are enough data points (Prophet requirement: >= 2 non-NaN rows)
if len(region_df) < 2:
    print(f"❌ ERROR: Region '{region.strip()}' has only {len(region_df)} non-NaN rows. Prophet requires at least 2.")
else:
    # Prophet expects columns: ds (date), y (value)
    df_prophet = region_df.rename(columns={"Date": "ds", "ZHVI_AllHomes": "y"})

    # Add regressors
    model = Prophet(yearly_seasonality=True)
    model.add_regressor("ZHVI_SingleFamilyResidence")
    model.add_regressor("ZHVIPerSqft_AllHomes")
    model.add_regressor("ZHVI_3bedroom")
    model.add_regressor("ZHVI_4bedroom")
    model.add_regressor("ZHVI_MiddleTier")
    model.add_regressor("ZHVI_TopTier")
    model.add_regressor("ZHVI_BottomTier")
    

    # Create and train model
    model = Prophet(daily_seasonality=False, yearly_seasonality=True)
    model.fit(df_prophet)

    # Forecast 24 months into the future (MS = Month Start frequency)
    future = model.make_future_dataframe(periods=72, freq="MS")

    # Add future values of regressors (for now use last known or mean value)
    future["ZHVI_SingleFamilyResidence"] = region_df["ZHVI_SingleFamilyResidence"].iloc[-1]
    future["ZHVIPerSqft_AllHomes"] = region_df["ZHVIPerSqft_AllHomes"].iloc[-1]
    future["ZHVI_3bedroom"] = region_df["ZHVI_3bedroom"].iloc[-1]
    future["ZHVI_4bedroom"] = region_df["ZHVI_4bedroom"].iloc[-1]
    future["ZHVI_MiddleTier"] = region_df["ZHVI_MiddleTier"].iloc[-1]
    future["ZHVI_TopTier"] = region_df["ZHVI_TopTier"].iloc[-1]
    future["ZHVI_BottomTier"] = region_df["ZHVI_BottomTier"].iloc[-1]

    # Forecast
    forecast = model.predict(future)

    # Plot forecast
    fig1 = model.plot(forecast)
    plt.title(f"Forecast of ZHVI_AllHomes for {region.strip()}")
    plt.xlabel("Year")
    plt.ylabel("ZHVI (USD)")
    plt.show()

    # Components (trend, yearly pattern)
    fig2 = model.plot_components(forecast)
    plt.show()
```

```python
import numpy as np
from sklearn.metrics import mean_absolute_error, mean_squared_error

# Evaluate on the last 24 rows of the region-specific dataframe used for Prophet
test = df_prophet.tail(72).reset_index(drop=True)

# Align forecast predictions with the test dates safely
pred_series = forecast.set_index("ds").reindex(test["ds"])["yhat"]

# Convert to numpy arrays and mask any missing predictions
y_true = test["y"].values
y_pred = pred_series.values

mask = ~np.isnan(y_pred)
if mask.sum() == 0:
    print("❌ No overlapping dates between forecast and test to compute metrics.")
else:
    mae = mean_absolute_error(y_true[mask], y_pred[mask])
    rmse = np.sqrt(mean_squared_error(y_true[mask], y_pred[mask]))
    
    # --- Add MAPE (Mean Absolute Percentage Error) ---
    mape = np.mean(np.abs((y_true[mask] - y_pred[mask]) / y_true[mask])) * 100
    
    # --- Convert to Accuracy Percentage ---
    accuracy = 100 - mape
    
    print(f"MAE: {mae:.2f}")
    print(f"RMSE: {rmse:.2f}")
    print(f"MAPE: {mape:.2f}%")
    print(f"Forecast Accuracy: {accuracy:.2f}%")
```

```python
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

plt.figure(figsize=(10,6))
plt.plot(test["ds"], y_true, label="Actual", marker='o', color='royalblue')
plt.plot(test["ds"], y_pred, label="Predicted", marker='x', color='darkorange')

# Format x-axis as readable dates
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m'))
plt.gca().xaxis.set_major_locator(mdates.AutoDateLocator())

plt.title("Actual vs Predicted Property Prices (Prophet Forecast)")
plt.xlabel("Date")
plt.ylabel("Price")
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```
